```python
from TTS.api import TTS  

# Charger un modèle TTS (tacotron2 pour une bonne qualité)
tts = TTS(model_name="tts_models/en/ljspeech/glow-tts", progress_bar=False, gpu=False)

# Générer un fichier audio
texte = "Yo, this is a test, ready to digest !"
tts.tts_to_file(text=texte, file_path="output.wav")

# Lire le fichier dans Jupyter
import IPython.display as ipd
ipd.Audio("output.wav")
```
